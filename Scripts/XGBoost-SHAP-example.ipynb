```python
import xgboost
import shap
from sklearn import tree
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df2016 = pd.read_csv('../Data/Final Data/data2016.csv')
df2016.head()
```

```python
# assign inputs and output values
X = df2016.iloc[:,9:]
y = df2016.iloc[:,8]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)

# Test Train split the data
X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, random_state=42)

from tensorflow.keras.utils import to_categorical
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
# load JS visualization code to notebook
shap.initjs()

model = xgboost.train({"learning_rate": 0.05, 'n_estimators':300, 'early_stopping_round':15}, xgboost.DMatrix(X_train, label=y_train))
# model = xgboost.train({'base_score':0.5, 'booster':'gbtree', 'colsample_bylevel':1,
#               'colsample_bynode':1, 'colsample_bytree':0.8, 'gamma':1.5,
#               'learning_rate':0.05, 'max_delta_step':0, 'max_depth':5,
#               'min_child_weight':1, 'missing':None, 'n_estimators':300, 'n_jobs':1,
#               'nthread':1, 'objective':'binary:logistic', 'random_state':0,
#               'reg_alpha':0, 'reg_lambda':1, 'scale_pos_weight':1, 
#               'silent':True, 'subsample':0.6, 'verbosity':1}, xgboost.DMatrix(X_train, label=y_train))

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_train)
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
plot_importance(model)
plt.show()
```

```python
# Single Force Plot
# Look at i =50, 25, 600
i = 50

shap.force_plot(explainer.expected_value, shap_values[i,:], X_train.iloc[i,:], matplotlib=True)
```

```python
# Force Plot
shap.force_plot(explainer.expected_value, shap_values, X_train)
```

```python
feature_shap = 'obesity'
shap.dependence_plot(feature_shap, shap_values, X_train, interaction_index=None)
```

```python
data2019 = pd.read_csv('../Data/Final Data/data2019.csv')

inputs = data2019.iloc[:,4:]

dtest = xgboost.DMatrix(inputs)

results = model.predict(dtest)
```

```python
# make predictions for test data
from sklearn.metrics import accuracy_score
X_test1 = xgboost.DMatrix(X_test)
y_pred = model.predict(X_test1)
predictions = [round(value) for value in y_pred]

# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
